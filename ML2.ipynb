```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 2: Load the dataset
df = pd.read_csv('50_Startups.csv')

# Step 3: Initialize OneHotEncoder with dense output
encoder = OneHotEncoder(sparse_output=False)  # Converts to a dense binary matrix

# Step 4: Fit and transform the "state" column
encoded_array = encoder.fit_transform(df[['State']])  # One-hot encode "state"

# Step 5: Create a DataFrame for the one-hot encoded columns
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['State']))

# Step 6: Merge the encoded columns with the original dataset
df_final = pd.concat([df.drop(columns=['State']), encoded_df], axis=1)

# Step 7: Display the final dataset
print(df_final.head())
```

```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 2: Load the dataset
df = pd.read_csv('50_Startups.csv')

# Step 3: Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # Ensures dense matrix output

# Step 4: Fit and transform the 'state' column
encoded_array = encoder.fit_transform(df[['State']])  # One-hot encode only the 'state' column

# Step 5: Convert the binary data into a DataFrame with meaningful column names
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['State']))

# Step 6: Concatenate the one-hot encoded column back with the original dataset
df_final = pd.concat([df.drop(columns=['State']), encoded_df], axis=1)

# Step 7: Display the final dataset
print(df_final)
 
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv('50_Startups.csv')

# Step 3: Define independent variables (X) and dependent variable (y)
X = df[['R&D Spend', 'Administration', 'Marketing Spend']]  # Independent variables
y = df['Profit']  # Dependent variable

# Step 4: Plot scatter plots for each independent variable vs Profit
plt.figure(figsize=(15, 5))  # Set the figure size

# Plot for R&D Spend vs Profit
plt.subplot(1, 3, 1)  # Create the first subplot
plt.scatter(X['R&D Spend'], y, color='blue')
plt.title('R&D Spend vs Profit')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')

# Plot for Administration vs Profit
plt.subplot(1, 3, 2)  # Create the second subplot
plt.scatter(X['Administration'], y, color='green')
plt.title('Administration vs Profit')
plt.xlabel('Administration')
plt.ylabel('Profit')

# Plot for Marketing Spend vs Profit
plt.subplot(1, 3, 3)  # Create the third subplot
plt.scatter(X['Marketing Spend'], y, color='red')
plt.title('Marketing Spend vs Profit')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')

# Show the plots
plt.tight_layout()  # Adjust the layout
plt.show()
```
