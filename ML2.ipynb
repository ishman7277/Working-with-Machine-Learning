```python
# Step 1: Import necessary libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 2: Load the dataset
df = pd.read_csv('50_Startups.csv')

# Step 3: Initialize OneHotEncoder with dense output
encoder = OneHotEncoder(sparse_output=False)  # Converts to a dense binary matrix

# Step 4: Fit and transform the "state" column
encoded_array = encoder.fit_transform(df[['State']])  # One-hot encode "state"

# Step 5: Create a DataFrame for the one-hot encoded columns
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['State']))

# Step 6: Merge the encoded columns with the original dataset
df_final = pd.concat([df.drop(columns=['State']), encoded_df], axis=1)

# Step 7: Display the final dataset
print(df_final.head())
```
